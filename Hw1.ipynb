```python
# (#5) Створіть два одновимірних масиви розміром 5, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та виконайте на них поелементні операції додавання, віднімання та множення.

# Розмірність масиву.
size = 5

# Створення двох масивів розміром 5, з випадковими цілими числами від 1 до 10.
array1 = np.random.randint(1, 11, size= size)
array2 = np.random.randint(1, 11, size=size)

print("Масив 1", array1)
print("Масив 2", array2)

# Поелементне додавання
sum_result = array1 + array2
print("Додавання", sum_result)

# Поелементне віднімання
subtract_result = array1 - array2
print("Віднімння", subtract_result)

# Поелементне множення
multiply_result = array1 * array2
print("Множення", multiply_result)
```

```python
# (#6) Створіть два вектори розміром 7, заповніть довільними числами та знайдіть їх скалярний добуток.

# Розмірність векторів
size = 7

# Створення двох векторів з довільними числами.
vector1 = np.random.rand(size)
vector2 = np.random.rand(size)

# Знаходження скалярного добутку.
dot_product = np.dot(vector1, vector2)

print("Вектор 1", vector1)
print("Вектор 2", vector2)
print("Скалярний добуток: ", dot_product)
```

```python
# (#8) Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть її обернену матрицю.

# # Створення випадкової матриці розміром 3x3 з цілими числами в діапазоні від 1 до 10.
random_matrix =np.random.randint(1, 11, size=(3, 3))

print("Випадкова матриця:")
print(random_matrix)

# Знаходження оберненої матриці.
inverse_matrix = np.linalg.inv(random_matrix)

print("Обернена матриця:")
print(inverse_matrix)
```

```python
# (#9) Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.

# Створення випадкової матриці розміром 4x4 з дійсними числами в діапазоні від 0 до 1.
random_matrix = np.random.rand(4, 4)

print("Випадкова матриця:")
print(random_matrix)

# Транспонування матриці.
transposed_matrix = np.transpose(random_matrix)
# Або коротший запис : transposed_matrix = random_matrix.T

print("Транспонована матриця:")
print(transposed_matrix)
```

```python
# (#10) Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та перемножте матрицю на вектор.

# Створення випадкової матриці розміром 3x4 з цілими числами в діапазоні від 1 до 10.
random_matrix = np.random.randint(1, 11, size=(3, 4))
print("Випадкова матриця:")
print(random_matrix)

# Створення випадкового вектора розміром 4 з цілими числами в діапазоні від 1 до 10.
random_vector = np.random.randint(1, 11, size=4)
print("Випадковий вектор:")
print(random_vector)

# Перемноження матриці на вектор.
result = np.dot(random_matrix, random_vector)
# Або коротший запис: result = random_matrix @ random_vector
print("Результат перемноження матриці на вектор:")
print(result)
```
